```python
import matplotlib.pyplot as plt

df['Total_Sales'].hist(figsize=(8,5))

plt.title('Sales Distribution')
plt.xlabel('Total Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
df['Category'].value_counts().plot(kind='bar', figsize=(8,5))

plt.title('Category Count')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

```python
top_products = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False)

print(top_products.head())
```

```python
city_sales = df.groupby('City')['Total_Sales'].sum().sort_values(ascending=False)

print(city_sales)
```

```python
numeric_df = df.select_dtypes(include='number')

correlation = numeric_df.corr()

print(correlation)
```
